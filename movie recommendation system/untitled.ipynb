Movie Recommendation System

The purpose of this project is to develop a machine learning-based movie recommendation system that provides personalized movie suggestions to users based on their preferences. By analyzing user ratings and movie attributes, the system aims to predict which films a user is most likely to enjoy, enhancing their overall experience.

To identify relevant movie recommendations, the system utilizes cosine similarity, a method that measures the similarity between movies. This technique compares the characteristics of movies a user has shown interest in with those of other films, ensuring that the recommendations are closely aligned with the user’s taste.

The required datasets for this project, including movie details, user ratings, and other relevant attributes, are imported in the form of CSV files sourced from the Kaggle website. These datasets provide the foundation for building the recommendation model, which processes and analyzes the data to generate accurate movie suggestions.

```python
# now we have to convert the list of dictionaries to list of strings

import ast # this python library is used to convert the string to list of dictionaries

def convert(text): # this function takes string as input and returns list as output
    L = [] # an empty list is created which will store the output
    
    for i in ast.literal_eval(text): # ast.literal_eval() converts the string to list of dictionaries
        # now we have to append the name of the dictionary to the list
        L.append(i['name'])
    
    return L # return the list
```

```python
movies.dropna(inplace=True) # remove the rows from 'movies' dataframe that contain missing values
```

```python
movies['genres'] = movies['genres'].apply(convert) # apply the function 'convert' to the 'genres' column of the dataframe
```

```python
movies.head() # testing if the function worked by checking values of column 'genres' of first 5 rows
```

as we see that the values of column 'genres' are easier to read as we extracted value of key 'name'
and created a list of the extracted values of each row.

```python
movies['keywords'] = movies['keywords'].apply(convert) # apply the 'convert' function on 'keywords' column
```

```python
movies.head() # checking if the function 'convert' worked on the dataframe by displaying it's first 5 rows
```

as we see that the function 'convert' successfully worked on column 'keywords' and now it is easier to read it's values as we simply extracted the values from the keys of dictionary formed using 'convert' function.

```python
movies['cast'] = movies['cast'].apply(convert) # similarly apply 'convert' function to 'cast' column
```

```python
movies.head() # testing if the function worked
```

as we see that the function worked and 'cast' column now includes only the cast name instead of other unnecessary information.

now the 'cast' column consists of list of many actors and actresses of the movie. But we need only three main casts of the movie. So we will modify the 'cast' column such that it contains list of only three main actors of the movie.

```python
movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
```

```python
movies.head() # now we will check if it worked by checking the first 5 rows of the dataframe
```

as we see that the lists of 'casts' column are trimmed to containing only the first 3 cast members.

```python
# create a function to fetch names of directors into a list

def fetch_director(text):
    L = [] # create an empty list in which we will append the names of directors 
    for i in ast.literal_eval(text): # evaluate the string
        if i['job'] == 'Director': # if the value of key 'job' is 'Director'
            L.append(i['name']) # append the value of key 'name' to the list
    return L # return the list formed
```

```python
movies['crew'] = movies['crew'].apply(fetch_director) # apply 'fetch_director' function on 'crew' column
```

```python
movies.head() # testing if the function 'fetch_director' works by seeing the first 5 rows of the dataset
```

as we see that the column 'crew' contains list of only director of the movie out of all crews

```python
# function to collapse a list of strings into a single string

def collapse(L): # this function takes a list 'L' as input
    L1 = [] # create an empty list 'L1' that will store the collapsed strings

    for i in L: # loop through each string in the input list 'L'
        L1.append(i.replace(" ","")) # remove any spaces from the current string and append it to 'L1'
    
    return L1
```

```python
# collapse the spaces in the string of cast, crew, genres, keywords columns

movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head() # testing if the function 'collapse' worked
```

as we see that the function 'collapse' worked and the values in lists of the four columns have no whitespace

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split()) # using lambda function to split the words in overview column into a list
# this means that column 'overview' is now a list of words instead of a string sepated by spaces
```

```python
movies.head() # testing if the splitting worked
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']

# create a column called 'tags' in the dataframe 'movies' by concatenating the columns 'overview', 'genres', 'keywords', 'cast', and 'crew'
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])

# create a dataframe 'new' which doesn't have columns 'overview','genres','keywords','cast','crew'
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x)) # the 'tags' column is a list of tags, we joined them into a string
# so now the column 'tags' now is a string created by concatenating all the values in the list separated by a space
```

```python
new.head() # testing if the previous code worked by printing the first 5 rows of the new dataframe
```

now that we have the data necessary, we can start to do some analysis using machine learning's cosine-similarity algorithm through Python's Scikit-Learn library.

```python
from sklearn.feature_extraction.text import CountVectorizer # 'CountVectorizer' is a class that will convert a collection of text documents to a matrix of counts of words.

cv = CountVectorizer(max_features=5000,stop_words='english') # 'cv' is an instance that will use only 5000 most frequent words and will ignore helping verbs in english like 'is' , 'are' , 'was' , 'were' , etc.

vector = cv.fit_transform(new['tags']).toarray() # from the string content of column 'tags' of dataframe 'new', create a matrix of counts of words and convert it to a 1-D array.
```

```python
from sklearn.metrics.pairwise import cosine_similarity # import the cosine similarity function from sklearn

similarity = cosine_similarity(vector)

# this function calculates the cosine similarity between the rows (movies) in 'vector' matrix
# 'vector' matrix contains word counts for each movie's tags (as produced by CountVectorizer)
# Cosine similarity measures how similar two vectors (movies) are, based on the cosine of the angle between them.
# A value of 1 means they are identical, 0 means they are orthogonal (completely different), and values in between indicate varying degrees of similarity.
```

```python
new[new['title'] == 'The Lego Movie'].index[0] # return index of first row of dataframe 'title' having name 'The Lego Movie'
```

```python
# create a function named 'recommend' that takes a movie title as input and returns a list of the top 5 recommended movies based on similarity using cosine similarity matrix.

def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
```

```python
recommend('Gandhi') # test the function 'recommend' by giving word 'Gandhi' as input
```

as we see that on the basis of similarity score matrix, we are able to find 5 most similar movies related to the word 'Gandhi'.
This includes movies containing the word 'Gandhi' in their title or any kind of similarity in value of any other column.
